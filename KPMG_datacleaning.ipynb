```python
import pandas as pd
import numpy as np

path = r"/Users/oksana_kaledina/Documents/5_PORTFOLIO/KPMG/data.xlsx"
df_transactions = pd.read_excel(path, sheet_name=1)
df_new_cust_list = pd.read_excel(path, sheet_name=2)
df_demographic = pd.read_excel(path, sheet_name=3) #clean
df_address = pd.read_excel(path, sheet_name=4) #clean

# index from 1
df_transactions.index = df_transactions.index + 1
df_new_cust_list.index = df_new_cust_list.index + 1
df_demographic.index = df_demographic.index + 1
df_address.index = df_address.index + 1
```

# Demographic

```python
# drop columns
df_demographic = df_demographic.drop(columns="default")
# DOB
df_demographic['DOB'] = pd.to_datetime(df_demographic['DOB']).dt.date
df_demographic = df_demographic.rename(columns={"DOB": "date_of_birth"})
# 'gender'
df_demographic['gender'] = df_demographic['gender'].str.replace('Female','F')
df_demographic['gender'] = df_demographic['gender'].str.replace('Femal','F')
df_demographic['gender'] = df_demographic['gender'].str.replace('Male','M')
df_demographic['gender'] = df_demographic['gender'].str.replace('F','Female')
df_demographic['gender'] = df_demographic['gender'].str.replace('M','Male')
# 'deceased_indicator'
df_demographic['deceased_indicator'] = df_demographic['deceased_indicator'].str.replace('N','No')
df_demographic['deceased_indicator'] = df_demographic['deceased_indicator'].str.replace('Y','Yes')
# null to ''
#df_demographic = df_demographic.fillna('NaN')
```

```python
df = df_address
df
```

```python
df.isnull().sum()
```

```python
# how many total missing values do we have?
missing_values_count = df.isnull().sum()
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()
percent_missing = (total_missing/total_cells) * 100

total_cells,total_missing, percent_missing.round(2)
```

```python
# dropna - columns
columns_with_na_dropped = df.dropna(axis=1)
print("Columns in original dataset: %d " % df.shape[1])
print("Columns with na's dropped: %d " % columns_with_na_dropped.shape[1])
a = df.shape[1] - columns_with_na_dropped.shape[1]
per_cent = a/df.shape[1]*100
print("Columns with na: %d "% a )
print("Percent of columns with na: %d"% per_cent + '%\n')
# dropna - rows
rows_with_na_dropped = df.dropna()
print("Rows in original dataset: %d" % df.shape[0])
print("Rows with na's dropped: %d " % rows_with_na_dropped.shape[0])
a = df.shape[0] - rows_with_na_dropped.shape[0]
per_cent = a/df.shape[0]*100
print("Rows with na: %d "% a)
print("Percent of rows with na: %d"% per_cent + '%')
```
